```python
# ImportaÃ§Ã£o de bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# 1. Carregar a base
df = pd.read_csv("/content/ProfissÃ£o Cientista de Dados M29 - online_shoppers_intention.csv")

# 2. Selecionar variÃ¡veis do escopo
variaveis = [
    "Administrative",
    "Administrative_Duration",
    "Informational",
    "Informational_Duration",
    "ProductRelated",
    "ProductRelated_Duration"
]
df_escopo = df[variaveis].copy()

# 3. EstatÃ­sticas descritivas
print("ðŸ“Š EstatÃ­sticas descritivas:")
print(df_escopo.describe())

# 4. VerificaÃ§Ã£o de valores ausentes
print("\nðŸ”Ž Valores ausentes:")
print(df_escopo.isnull().sum())

# 5. DistribuiÃ§Ãµes
plt.figure(figsize=(12,8))
for i, col in enumerate(variaveis, 1):
    plt.subplot(2, 3, i)
    sns.histplot(df_escopo[col], bins=30, kde=True, color="royalblue")
    plt.title(f"DistribuiÃ§Ã£o de {col}")
plt.tight_layout()
plt.show()

# 6. PadronizaÃ§Ã£o
scaler = StandardScaler()
X = scaler.fit_transform(df_escopo)

# 7. MÃ©todo do Cotovelo
inercia = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inercia.append(kmeans.inertia_)

# 8. Silhouette Score
silhouette_scores = []
K2 = range(2, 11)
for k in K2:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    score = silhouette_score(X, kmeans.labels_)
    silhouette_scores.append(score)

# 9. GrÃ¡ficos lado a lado
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].plot(K, inercia, 'bo-')
axes[0].set_title("MÃ©todo do Cotovelo")
axes[0].set_xlabel("NÃºmero de clusters (k)")
axes[0].set_ylabel("InÃ©rcia")

axes[1].plot(K2, silhouette_scores, 'ro-')
axes[1].set_title("Coeficiente de Silhueta")
axes[1].set_xlabel("NÃºmero de clusters (k)")
axes[1].set_ylabel("Silhouette Score")

plt.tight_layout()
plt.show()

# 10. ClusterizaÃ§Ã£o final (k=4, melhor silhueta)
kmeans = KMeans(n_clusters=4, random_state=42)
df_escopo["Cluster"] = kmeans.fit_predict(X)

# 11. DistribuiÃ§Ã£o dos clusters
print("\nðŸ“Š DistribuiÃ§Ã£o dos clusters:")
print(df_escopo["Cluster"].value_counts())

# 12. MÃ©dias das variÃ¡veis por cluster
print("\nðŸ“Š MÃ©dias das variÃ¡veis por cluster:")
print(df_escopo.groupby("Cluster").mean())

# 13. AvaliaÃ§Ã£o com variÃ¡veis fora do escopo
print("\nðŸ“ˆ AvaliaÃ§Ã£o com BounceRates e Revenue:")
avaliacao = df.groupby(df_escopo["Cluster"])[["BounceRates","Revenue"]].mean()
print(avaliacao)

# 14. GrÃ¡fico de Revenue por cluster
plt.figure(figsize=(6,4))
sns.barplot(x=df_escopo["Cluster"], y=df["Revenue"].astype(int), estimator=np.mean, hue=df_escopo["Cluster"], palette="viridis", legend=False)
plt.title("ProporÃ§Ã£o de compras (Revenue) por cluster")
plt.xlabel("Cluster")
plt.ylabel("ProporÃ§Ã£o de compras")
plt.show()
```
